## Formula used in Simple Linear Regression to calculate slope (m) and intercept (c).

We fit the line:

$[
y = m x + c
]$

---

### Slope (m) formula

$[
m = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
]$

- Slope (m) = covariance(x, y) / variance(x)
---

### Intercept (c) formula

$[
c = \bar{y} - m \bar{x}
]$

Where:

* $( \bar{x} )$ = mean of x
* $( \bar{y} )$ = mean of y

- Intercept (c) = mean(y) − m × mean(x)
---




```python
class MySLR:

    def __init__(self):
        self.m = None
        self.c = None

    def fit(self, X_train,y_train):

        num = 0
        den = 0
        
        for i in range(X_train.shape[0]):

            num = num + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
            den = den + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))
            
        self.m = num / den
        self.c = y_train.mean() - (self.m * X_train.mean())
        print(self.m)
        print(self.c)

    def predict(self,X_test):
        return self.m * X_test + self.c
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv("placement.csv")
df.head()
```

```python
X = df.iloc[:,0].values # Input feature
y = df.iloc[:,-1].values # Output feature
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
# Model Building

lr = MySLR()
lr.fit(X_train, y_train)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data = df, x = "cgpa", y = "package")
plt.plot(X_train, lr.predict(X_train), color ="red")
plt.xlabel("cgpa")
plt.ylabel("package")
```

```python
y_pred = lr.predict(X_test) # Prediction
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error = ",mse)
print("Mean Absolute Error = ",mae)
print("Root Mean Squared Error = ",rmse)
print("R2 Score = ",r2)
```
